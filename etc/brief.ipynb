```py
def factorial(n):
    # ì¢…ë£Œ ì¡°ê±´: nì´ 0ì´ë©´ 1ì„ ë°˜í™˜
    if n == 0:
        return 1
    else:
        # ì¬ê·€ í˜¸ì¶œ: nê³¼ n-1ì˜ íŒ©í† ë¦¬ì–¼ì„ ê³±í•œ ê²°ê³¼ë¥¼ ë°˜í™˜
        return n * factorial(n - 1)


# íŒ©í† ë¦¬ì–¼ ê³„ì‚° ì˜ˆì‹œ
print(factorial(5))  # 120
```

$5! = 5 \cdot 4 \cdot 3 \cdot 2 \cdot 1$
