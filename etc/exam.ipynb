```python
# 재귀함수의 필수 구성 요소
# 1. 시작점, 누적값
# 2. 종료조건

arr = [2, 4, 6, 8, 10]

# 2 4 6 8 10 8 6 4 2

# 2 4 6 8 10
def func(idx):

    if idx == len(arr) - 1:
        # 종료 조건이 만족했을 때, 수행되어야 할 로직
        print(arr[idx], end=' ') 
        return
    
    print(arr[idx], end=' ') # 다음 재귀 호출 전에 뭘 하고 싶나?

    func(idx + 1)   # 다음 재귀 호출에는 어떤 값을 전댈해줘야 하나?
    print(arr[idx], end=' ') # 돌아오면서는 어떤 로직을 수행하고 싶나?

func(0)

# func(0) -> func(1) -> func(2) -> func(3)
```

```python
# 십진수 -> 2진수 변환 (재귀호출)
# 10 -> 1010

# 1. 2로 나눔
# 2. 결과에 나머지 값 추가

def binary(num):
    if num < 2:
        print(num, end='')
        return
    
    binary(num//2)
    print(num%2, end='')

binary(10)
```

```python
# 십진수 -> 2진수 변환 (재귀호출)
# 10 -> 1010

# 1. 2로 나눔
# 2. 결과에 나머지 값 추가

def binary(num):
    if num < 2:
        print(num,end='')
        return

    binary(num//2)
    print(num%2,end='')


binary(10)
```

```python
# True False 문제
item1 = '지도'
item2 = '나침반'

print(item1 and item2)  # True
print(item1 or item2)   # True
print("-----")
item1 = '지도'
item2 = ''

print(item1 and item2) # False
print(item1 or item2)  # True
print("-----")

item1 = ''
item2 = '지도'

print(item1 and item2) # False
print(item1 or item2)  # True
print((item1 and item2) == False)
print((item1 and item2) == '')
print('' == False)
print("-----")
if '':  # 조건문에선 False로 인식
    print("값이 비어있다.")

if not 0:   # not 0은 True이기에 출력됨
    print('된다')
```

### 해싱(Hashing)



key -> hash function -> bucket

만약 키값이 880 -> return 880 % bucket_size

해시 충돌 : 다른 키인데 저장 공간이 이미 차있는 곳으로 나옴 -> 다음 칸에 넣음
버킷이 꽉 차면 사이즈를 2배로 늘림

```python
# 해시함수 구현하기
bucket = [0] * 8 # [0, 0, 0, 0, 0, 0, 0, 0]
bucket_size = 8
# 충돌이 너무 많이 발생하는, 안좋은 해시 함수
# 나중에는 충돌이 없도록 해시함수를 구현해야 한다.
# 해시충돌 : 다른 key값이 같은 hash 값을 가지는 경우
def simple_hash(word):
    # a=1, b=2, c=3, ..., z=26
    word = word.lower() # 모두 소문자로 처리할 것
    hash_value = 0

    for char in word:
        hash_value += ord(char) - 96

    return hash_value % bucket_size

simple_hash('abc')
simple_hash('bbb')
simple_hash('ccc')
```

```python
# ASCII code
# ord() : 아스키 코드로 변환하여 출력
print(ord("a"))
print(ord("b"))
# chr() : 아스키 코드 -> 문자로 변환
print(chr(97))

# 소문자 a~z를 순서대로 출력하라
for w in range(ord('a'),ord('z')+1):
    print(chr(w))
```
